**Problem Identification**: Sales Prediction Using Machine Learning` <br />
**Input**: TV, Radio, Newspaper <br />
**Output**: Sales <br />
**Dataset**: Advertising.csv <br />

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('advertising.csv')
print(data)
print(data.shape)
```

```python
train, test = train_test_split(data, test_size = 0.2)
print(train.shape)
print(test.shape)

X_train = train[['TV', 'Radio', 'Newspaper']]

# print(X_train.shape)

y_train = train.Sales

# print(y_train.shape)

X_test = test[['TV', 'Radio', 'Newspaper']]
y_test = test.Sales

```

```python
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)
```

```python
y_pred = rf_regressor.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared score: {r2}')
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual Sales vs. Predicted Sales (Random Forest)")
plt.show()
```
